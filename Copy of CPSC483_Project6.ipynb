###Project Summary 
This project requires the creation of a KNN, Naive Bayes, and SVM for bank data collected by UCI, the goal is to predict whether they said yes or not. The point is to semonstrate that sometimes due to there being to many of one class versus the other it is nessecary to resample the data to minimize the effect of this


```python
import numpy as np
import pandas as pd
from sklearn import naive_bayes, neighbors, svm, metrics
from imblearn.over_sampling import RandomOverSampler
```

### Experiment 2

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python

training_set = pd.read_csv("/content/drive/MyDrive/bank-additional/bank-additional.csv", delimiter = ';')
test_set = pd.read_csv("/content/drive/MyDrive/bank-additional/bank-additional-full.csv", delimiter = ';')
```

### Experiment 3a

```python
test_set = test_set.dropna()

training_set = pd.get_dummies(training_set, drop_first= True)
test_set = pd.get_dummies(test_set, drop_first= True)
```

### Experiment 3b

```python
if 'duration' in training_set.columns:
    training_set.pop('duration')
if 'duration' in test_set.columns:
    test_set.pop('duration')
```

```python

if 'y_yes' in training_set.columns:
    training_target = training_set.pop('y_yes')
if 'y_yes' in test_set.columns:
    test_target = test_set.pop('y_yes')
```

```python
test_target.shape
```

```python
test_set.shape
```

### Experiment 3c

```python
to_normalize = ['age', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

for label in to_normalize:
    mean =  training_set[label].mean()
    std = training_set[label].std()
    print(std)
    training_set[label] = (training_set[label] - mean)/std
    test_set[label] = (test_set[label] - mean)/std
```

### Experiment 4

```python
gnb = naive_bayes.GaussianNB().fit(training_set, training_target)
gnb_training_score = gnb.score(training_set, training_target)
gnb_test_score = gnb.score(test_set, test_target)
print('Naive Bayes training score = ', gnb_training_score)
print("Naive Bayes test score = ", gnb_test_score)
```

```python
knn = neighbors.KNeighborsClassifier().fit(training_set, training_target)
knn_training_score = knn.score(training_set, training_target)
knn_test_score = knn.score(test_set, test_target)
print('KNN training score = ', knn_training_score)
print("KNN test score = ", knn_test_score)
```

```python
suportvm = svm.SVC().fit(training_set, training_target)
svm_training_score = suportvm.score(training_set, training_target)
svm_test_score = suportvm.score(test_set, test_target)
print('SVM training score = ', svm_training_score)
print("SVM test score = ", svm_test_score)
```

The highest accuracy is the Support Vector Machine

### Experiment 8

```python
ros =  RandomOverSampler(random_state=(2021-4-22))
rs_training_set, rs_training_target = ros.fit_resample(training_set, training_target)
rs_test_set, rs_test_target = ros.fit_resample(test_set, test_target)
```

### Eperiment 9

```python
gnb2 = naive_bayes.GaussianNB().fit(rs_training_set, rs_training_target)
gnb_training_score2 = gnb2.score(rs_training_set, rs_training_target)
gnb_test_score2 = gnb2.score(rs_test_set, rs_test_target)
print('Naive Bayes2 training score = ', gnb_training_score2)
print("Naive Bayes2 test score = ", gnb_test_score2)
```

```python
knn2 = neighbors.KNeighborsClassifier().fit(rs_training_set, rs_training_target)
knn_training_score2 = knn2.score(rs_training_set, rs_training_target)
knn_test_score2 = knn2.score(rs_test_set, rs_test_target)
print('KNN training score2 = ', knn_training_score2)
print("KNN test score = ", knn_test_score2)
```

```python
suportvm2 = svm.SVC().fit(rs_training_set, rs_training_target)
svm_training_score2 = suportvm2.score(rs_training_set, rs_training_target)
svm_test_score2 = suportvm2.score(rs_test_set, rs_test_target)
print('SVM training score = ', svm_training_score2)
print("SVM test score = ", svm_test_score2)
```

```python
gnb_pred2 = gnb2.predict(rs_test_set)
metrics.confusion_matrix(rs_test_target, gnb_pred2)
```

```python
metrics.roc_auc_score(rs_test_target,gnb_pred2)
```

```python
knn_pred2 = knn2.predict(rs_test_set)
metrics.confusion_matrix(rs_test_target, knn_pred2)
```

```python
metrics.roc_auc_score(rs_test_target, knn_pred2)
```

```python
svm_pred2 = suportvm.predict(rs_test_set)
metrics.confusion_matrix(rs_test_target, svm_pred2)
```

```python
metrics.roc_auc_score(rs_test_target, svm_pred2)
```
