# Expirment 2


```
test =[]
training = []
message = []
for image, letter in TEST_SET:

    test.append((np.array(image), letter))

for image, letter in TRAINING_SET:
  
    training.append((np.array(image), letter))

for image in MESSAGE:
    message.append(np.array(image))

message = np.array(message)
```

```
def show(image):
    collum_count = 0
    for i in image:
        if i == 1:
            print('#' , end = '')
        else:
            print(' ', end='')
        collum_count +=1
        if collum_count == 5:
            print("")
            collum_count = 0
    print('\n')
```

```
for i in training:
    show(i[0])
```

#Experiment 3

```

temp =[]
test_cat = []
for image, letter in test:
    temp.append(image)
    test_cat.append((ord(letter) - 65) )
test_set = np.array(temp)
test_cat = keras.utils.to_categorical(test_cat)

temp =[]
training_cat = []
for image, letter in training:
    temp.append(image)
    training_cat.append(ord(letter) - 65)
training_set = np.array(temp)
training_cat = keras.utils.to_categorical(training_cat)

```

```




print(len(training_cat))
print(type(training_cat))
print(training_cat.shape)
print(training_cat[0])
print(training_cat[0][0])
```

#Experiment 4

```
model = keras.Sequential()
model.add(keras.Input(shape =(35,), name='Input' ))
model.add(keras.layers.Dense(26, activation= 'softmax', name= 'hidden'))
model.add(keras.layers.Dense(26, activation= 'softmax', name= 'output'))
model.summary()
```

#Experiment 5

```
dataset = tf.data.Dataset.from_tensor_slices((training_set, training_cat)).batch(52)
opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer = opt, loss= 'categorical_crossentropy',  metrics = [tf.keras.metrics.CategoricalAccuracy()]  )
history = model.fit(dataset, epochs= 10000, callbacks= [keras.callbacks.EarlyStopping(monitor='loss', min_delta=0, patience= 5)])
```

#Experiment 6

```
loss, acc =model.evaluate(x =test_set, y= test_cat, batch_size= 26)
```

```
wrong =[]
for data, label in zip( test_set, test_cat):
    
    loss, acc = model.evaluate(x= np.array([data]), y = np.array([label]), verbose= 0)
    if acc == 0:
        wrong.append(chr(65+(np.argmax(label, axis= -1))))
print("The following were misclasified")
print(wrong)
```

#Experiment 7

```

predicted = model.predict(x=message)

for i in predicted:
    print(chr(65+(np.argmax(i, axis= -1))), end='')
```

###The predictions were a little better for the messages but still pretty bad
